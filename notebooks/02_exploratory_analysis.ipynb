# Customer Lifecycle Analytics  
## Exploratory Data Analysis (EDA)

### Objective
The objective of this notebook is to explore customer churn patterns and
identify behavioral, demographic, and engagement-related factors that may
influence customer exit decisions.

### Key Business Questions
- What percentage of customers are churning?
- Does churn vary by geography or gender?
- How do age, tenure, and balance relate to churn?
- Do engagement indicators such as activity status, complaints, and
  satisfaction scores influence churn?

### Dataset Used
This notebook uses the cleaned dataset produced in the data understanding
stage:



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/customer_churn_clean.csv")
```

```python
churn_counts = df["churn"].value_counts()
churn_percent = df["churn"].value_counts(normalize=True) * 100

churn_counts, churn_percent
```

```python
geo_churn = pd.crosstab(df["geography"], df["churn"], normalize="index") * 100
geo_churn
```

```python
geo_churn.plot(kind="bar", stacked=True)
plt.title("Churn Percentage by Geography")
plt.ylabel("Percentage")
plt.show()
```

```python
gender_churn = pd.crosstab(df["gender"], df["churn"], normalize="index") * 100
gender_churn
```

```python
gender_churn.plot(kind="bar", stacked=True)
plt.title("Churn Percentage by Gender")
plt.ylabel("Percentage")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="churn", y="age", data=df)
plt.title("Age Distribution by Churn Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="churn", y="tenure", data=df)
plt.title("Tenure Distribution by Churn Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="churn", y="balance", data=df)
plt.title("Account Balance by Churn Status")
plt.show()
```

```python
product_churn = pd.crosstab(df["numofproducts"], df["churn"], normalize="index") * 100
product_churn
```

```python
product_churn.plot(kind="bar", stacked=True)
plt.title("Churn by Number of Products")
plt.ylabel("Percentage")
plt.show()
```

```python
active_churn = pd.crosstab(df["is_active_member"], df["churn"], normalize="index") * 100
active_churn
```

```python
active_churn.plot(kind="bar", stacked=True)
plt.title("Churn by Active Membership Status")
plt.ylabel("Percentage")
plt.show()
```

```python
complain_churn = pd.crosstab(df["complain"], df["churn"], normalize="index") * 100
complain_churn
```

```python
complain_churn.plot(kind="bar", stacked=True)
plt.title("Churn by Complaint Status")
plt.ylabel("Percentage")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="churn", y="satisfaction_score", data=df)
plt.title("Satisfaction Score by Churn Status")
plt.show()
```
